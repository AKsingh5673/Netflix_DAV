```python
# visualising the number of Movies and TV Shows in the dataset
plt.subplots(figsize=(15,5))
data['Type'].value_counts(normalize = True)
data['Type'].value_counts(dropna = False).plot.bar(color = ['purple','blue'])
plt.title('Comparison of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
data['Country'].value_counts()
```

```python
# visualizing the different Rating in the dataset
df = data.groupby('Country').size()
df.plot(kind='pie', subplots=True, figsize=(10, 8))
plt.title("Pie Chart of Country")
plt.ylabel("")
plt.show()
```

```python
# visualizing the differnt Rating
import matplotlib.pyplot as plt
plt.subplots(figsize=(18,5))
data['Country'].value_counts(normalize = True)
data['Country'].value_counts(dropna = False).plot.bar(color=['black', 'red', 'green', 'blue', 'cyan'])
plt.title('Comparison of Rating')
plt.xlabel('DRating')
plt.ylabel('Count')
plt.show()
```

```python
# visualizing the differnt Rating in pie chart
df = data.groupby('Rating').size()
df.plot(kind='pie', subplots=True, figsize=(10, 8))
plt.title("Pie Chart of Rating")
plt.ylabel("")
plt.show()
```

```python
# visualizing the differnt Rating
import matplotlib.pyplot as plt
plt.subplots(figsize=(18,5))
data['Rating'].value_counts(normalize = True)
data['Rating'].value_counts(dropna = False).plot.bar(color=['black', 'red', 'green', 'blue', 'cyan'])
plt.title('Comparison of Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# visualizing comparision of Date
plt.subplots(figsize=(15,5))
data['Date'].value_counts(normalize = True)
data['Date'].value_counts(dropna = False).plot.bar(figsize = (18, 10), color = 'orange')
plt.title('Comparison of Released Date')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
# visualizing comparision of Years
plt.subplots(figsize=(15,5))
data['Year'].value_counts(normalize = True)

data['Year'].value_counts(dropna = False).plot.bar(figsize = (18, 10), color = 'blue')
plt.title('Comparison of Years')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
# visualizing comparision of Years in Histogram
plt.subplots(figsize=(15,5))
plt.hist(data['Year'], bins=20, color='cyan')
plt.title('Comparison of Years')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```
